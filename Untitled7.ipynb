```python
`from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Myapp").getOrCreate()
```

```python
df1 = spark.read.csv("/content/Category.csv",header=True,inferSchema=True)
df2 = spark.read.csv("/content/Customers.csv",header=True,inferSchema=True)
df3 = spark.read.csv("/content/OrderDetails.csv",header=True,inferSchema=True)
df4 = spark.read.csv("/content/Orders.csv",header=True,inferSchema=True)
df5 = spark.read.csv("/content/Payments.csv",header=True,inferSchema=True)
df6 = spark.read.csv("/content/Products.csv",header=True,inferSchema=True)
df7 = spark.read.csv("/content/Shippers.csv",header=True,inferSchema=True)
df8 = spark.read.csv("/content/Suppliers.csv",header=True,inferSchema=True)
```

```python
sales = df4.join(df2,on="CustomerID",how="left").drop(df2.CustomerID)\
       .join(df3,on="OrderID",how="left")\
       .join(df6,on="ProductID",how="left").drop(df6.ProductID)
sales.show()
```

```python
sales.printSchema()
```

```python
from pyspark.sql.functions import col, sum, round,desc

sales = sales.withColumn("Total_sale", col("Sale_Price") * col("Quantity"))

sale = sales.groupBy("City") \
    .agg(round(sum("Total_sale"), 2).alias("Total_sale")) \
    .orderBy(desc("Total_sale")) \
    .limit(5)

sale.show()
```

```python
sales.printSchema()
```

```python
sales=sales.join(df5,on="PaymentID",how="left").drop(df5.PaymentID)
sales.show()
```

```python
sales = sales.join(df1, on=sales.Category_ID == df1.CategoryID, how="left").drop(df1.CategoryID)
sales.show()
```

```python
from pyspark.sql.functions import col, sum, round,desc
#sales = sales.withColumn("Total_sales_price",col("Market_Price")*col("Quantity"))
result = sales.groupBy("PaymentType", "CategoryName") \
    .agg(round(sum("Total_order_amount"),2).alias("Total_sales")).orderBy(desc("Total_sales"))


result.show()
```

```python
from pyspark.sql.functions import col, sum, round,desc
#sales = sales.withColumn("Total_sales_price",col("Market_Price")*col("Quantity"))
sales = sales.withColumn("Profit",col("Sale_Price")-col("Market_Price"))

profit = sales.filter(sales.Country == "United States") \
              .groupBy(col("CategoryName")) \
              .agg(sum(col("Profit")).alias("TotalProfit"))\
              .orderBy(desc("TotalProfit"))

profit.show()
```

```python
from pyspark.sql.functions import col, sum, round, asc

no_of_sale = sales.groupBy("Category_ID", "CategoryName").agg(sum("Quantity").alias("Total_sold")).orderBy(asc("Total_sold"))

no_of_sale.show()
```

```python
sales.printSchema()
```

```python
from pyspark.sql.functions import col, round, sum, avg, desc

sub = sales.groupBy("Sub_Category").agg(
    round(sum("Quantity"), 2).alias("Quantity_sold"),
    avg("Rating").alias("Avg_rating")).orderBy(desc("Quantity_sold"))

sub.show()
```

```python
from pyspark.sql.functions import to_date,datediff,col
sales = sales.withColumn("ShipDate",to_date(col("ShipDate"),"dd-MM-yyyy"))
sales = sales.withColumn("DeliveryDate",to_date(col("DeliveryDate"),"dd-MM-yyyy"))
sales = sales.withColumn("DeliveryDays",datediff(col("DeliveryDate"),col("ShipDate")))
sales.printSchema()
```

```python
sales = sales.join(df7,on="ShipperID",how="inner")
```

```python
sales.printSchema()
```

```python
from pyspark.sql.functions import col, count, avg, round, desc, asc

ship = sales.groupBy("ShipperID", "CompanyName")\
.agg(round(avg("DeliveryDays"),2).alias("AvgDeliveryTime"))\
.orderBy("AvgDeliveryTime" , ascending=False)
ship.show()
```

```python
from pyspark.sql.functions import count ,round
per = sales.groupBy("CustomerID").agg(count("OrderID").alias("OrderCount"))
repeat_customers = per.filter("OrderCount > 1")
total_customers = per.select("CustomerID").distinct().count()
repeat_customers_count = repeat_customers.count()
retention_rate =(repeat_customers_count / total_customers) * 100
print("Customer Retention Rate:", retention_rate, "%")
```

```python
from pyspark.sql.functions import *
john = sales.filter(sales.FirstName=="John").agg(count("Quantity").alias("Total_sold")).show()
```
