```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("test").getOrCreate()
```

```python
df1 = spark.read.csv("/content/Category.csv",header=True,inferSchema=True)
df2 = spark.read.csv("/content/Customers.csv",header=True,inferSchema=True)
df3 = spark.read.csv("/content/OrderDetails.csv",header=True,inferSchema=True)
df4 = spark.read.csv("/content/Orders.csv",header=True,inferSchema=True)
df5 = spark.read.csv("/content/Payments.csv",header=True,inferSchema=True)
df6 = spark.read.csv("/content/Products.csv",header=True,inferSchema=True)
df7 = spark.read.csv("/content/Shippers.csv",header=True,inferSchema=True)
df8 = spark.read.csv("/content/Suppliers.csv",header=True,inferSchema=True)
```

```python
df4.show()
```

```python
df8.printSchema()
```

```python
sales = df4.join(df2,on="CustomerID",how="left")
sales.show()
```

```python
from pyspark.sql.functions import sum,desc,round
top_5_city = sales.groupBy("City").agg(round(sum("Total_order_amount"),2).alias("Total_sales")).orderBy(desc("Total_sales")).limit(5).show()
```

```python
sales=sales.join(df5,on="PaymentID",how="left")\
.join(df3,on="OrderID",how="left")\
.join(df6,on="ProductID",how="left")
sales.printSchema()
```

```python
sales = sales.join(df1,on = sales.Category_ID == df1.CategoryID,how="left")
sales.show()
```

```python
sales.printSchema()
```

```python
from pyspark.sql.functions import sum,desc,round,format_number
most_revenue = sales.groupBy("PaymentType","CategoryName").agg(round(sum("Total_order_amount"),2).alias("Total_sales"))\
.orderBy(desc("Total_sales"))
most_revenue.show()
```

```python
from pyspark.sql.functions import sum,desc,round,col

sales = sales.withColumn("Profit",col("Sale_Price")-col("Market_Price"))
top_sale_us = sales.filter(sales.Country == "United States").groupBy("CategoryName").agg(round(sum("Profit"),2).alias("Total_profit"))\
.orderBy(desc("Total_profit")).show()
```

```python
sales.printSchema()
```

```python
from pyspark.sql.functions import sum,desc,avg,round

avg_rating = sales.groupBy("Sub_Category").agg(sum("Quantity").alias("TotalQuantitySold"),avg("Rating").alias("Avg_Rating"))\
.orderBy(desc("TotalQuantitySold")).show()
```

```python
from pyspark.sql.functions import count
rdd3 = sales.groupBy("CustomerID").agg(count("OrderID").alias("OrderCount"))
repeat_customers = rdd3.filter("OrderCount > 1")
total_customers = rdd3.select("CustomerID").distinct().count()
repeat_customers_count = repeat_customers.count()
retention_rate = (repeat_customers_count / total_customers) * 100
print("Customer Retention Rate:", retention_rate, "%")
```

```python
from pyspark.sql.functions import count, sum, avg,lit


customer_orders = sales \
    .groupBy("CustomerID") \
    .agg(
        count("OrderID").alias("OrderCount"),
        sum("Total_order_amount").alias("TotalSpent"),
        avg("Total_order_amount").alias("AvgOrderValue")
    )
one_time_customers = customer_orders.filter("OrderCount = 1")
repeat_customers = customer_orders.filter("OrderCount > 1")

summary_one_time = one_time_customers.agg(
    count("CustomerID").alias("NumCustomers"),\
    avg("OrderCount").alias("AvgOrderCount"),\
    avg("TotalSpent").alias("AvgTotalSpent"),\
    avg("AvgOrderValue").alias("AvgOrderValue")\
).withColumn("CustomerType", lit("One-Time Customer"))

summary_repeat = repeat_customers.agg( \
    count("CustomerID").alias("NumCustomers"),\
    avg("OrderCount").alias("AvgOrderCount"),\
    avg("TotalSpent").alias("AvgTotalSpent"),\
    avg("AvgOrderValue").alias("AvgOrderValue")\
).withColumn("CustomerType", lit("Repeat Customer"))

final_summary = summary_one_time.union(summary_repeat)
final_summary.select("CustomerType", "NumCustomers", "AvgOrderCount", "AvgTotalSpent", "AvgOrderValue").show(truncate=False)
```

```python
from pyspark.sql.functions import to_date,datediff,col

sales = sales.withColumn("ShipDate",to_date(col("ShipDate"),"dd-MM-yyyy"))
sales = sales.withColumn("DeliveryDate",to_date(col("DeliveryDate"),"dd-MM-yyyy"))
sales = sales.withColumn("DeliveryDay",datediff(col("DeliveryDate"),col("ShipDate")))
sales.printSchema()
```
